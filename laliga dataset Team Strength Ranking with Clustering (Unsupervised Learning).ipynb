```python
import pandas as pd
```

```python
df = df= pd.read_csv('D:\datasets to publish\LaLiga Matches Dataset (2019-2025, FBref)\matches_full.csv')
df
```

```python
cols_to_keep = [
    "team","poss","sh","sot","fk","pkatt","xg","xga","dist"
]
df = df[cols_to_keep].copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['team'] = le.fit_transform(df['team'])
```

```python
team_statues = df.groupby("team").agg({
    "poss":"mean",
    "sh":"mean",
    "sot" : "mean",
    "dist" : "mean",
    "xg" : "mean",
    "xga":"mean",
    "fk" : "mean",
    "pkatt" : "mean"
    
}).reset_index()
```

```python
team_statues
```

```python
fetaures_cols = [
    "poss","sh","sot","dist","xg","xga",'fk',"pkatt"
]
x = team_statues[fetaures_cols].values
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x = scale.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
inertias,silhouettes =[],[]
k_range = range(2,10)
for k in k_range:
    model = KMeans(n_clusters=k,random_state=42,n_init=10)
    labels =model.fit_predict(x)
    inertias.append(model.inertia_)
    silhouettes.append(silhouette_score(x,labels))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.Figure(figsize=(10,5))
plt.plot(k_range,inertias,"o-")
plt.title("Elbow Method")
plt.xlabel(("number of clusters (k)"))
plt.ylabel(inertias)
plt.grid(True)
plt.show()
```

```python
best_k=3
kmeans = KMeans(n_clusters=best_k, random_state=42,n_init=10)
team_statues["cluster"] =kmeans.fit_predict(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)
plt.figure(figsize=(8,6))
pallette = sns.color_palette("Set2",best_k)
for i in range(best_k):
    plt.scatter(
        x_pca[team_statues["cluster"] == i,0],
        x_pca[team_statues["cluster"] == i,1],
        s=100,
        label = f"cluster {i}",
        color = pallette [i]
    )
    team_name = le.inverse_transform(team_statues["team"])
    plt.title ("La liga team Clusters (PCA PRojection)")
    plt.xlabel("PC1")
    plt.ylabel("PCA2")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
cluster_profiles = team_statues.groupby("cluster")[fetaures_cols].mean().round(2)
print("\n cluser profiles :")
display(cluster_profiles)
```
