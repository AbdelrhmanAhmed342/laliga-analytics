```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Assuming your dataframe is named 'df'
# Let's select numeric columns to transform (excluding binary/categorical columns)
numeric_cols = ['gf', 'xg', 'poss', 'sh', 'sot', 'dist', 'attendance']

# Function to plot distribution before and after transformation
def plot_transformation(original, transformed, title):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(original, kde=True, ax=axes[0])
    axes[0].set_title(f'Original {title}')
    sns.histplot(transformed, kde=True, ax=axes[1])
    axes[1].set_title(f'Transformed {title}')
    plt.tight_layout()
    plt.show()

# 1. Log Transformation (add 1 to handle zeros)
for col in numeric_cols:
    if df[col].min() > 0:  # Log works only on positive values
        df[f'log_{col}'] = np.log(df[col])
        plot_transformation(df[col], df[f'log_{col}'], f'Log {col}')
    else:
        # Add minimum value + 1 to make all values positive
        shift = abs(df[col].min()) + 1
        df[f'log_{col}'] = np.log(df[col] + shift)
        plot_transformation(df[col], df[f'log_{col}'], f'Log ({col}+{shift})')

# 2. Square Root Transformation
for col in numeric_cols:
    if df[col].min() >= 0:  # Square root works on non-negative values
        df[f'sqrt_{col}'] = np.sqrt(df[col])
        plot_transformation(df[col], df[f'sqrt_{col}'], f'Sqrt {col}')
    else:
        # Shift data to make all values non-negative
        shift = abs(df[col].min())
        df[f'sqrt_{col}'] = np.sqrt(df[col] + shift)
        plot_transformation(df[col], df[f'sqrt_{col}'], f'Sqrt ({col}+{shift})')

# 3. Box-Cox Transformation (more sophisticated)
for col in numeric_cols:
    # Box-Cox requires strictly positive values
    if df[col].min() > 0:
        try:
            transformed, _ = stats.boxcox(df[col])
            df[f'boxcox_{col}'] = transformed
            plot_transformation(df[col], df[f'boxcox_{col}'], f'Box-Cox {col}')
        except ValueError as e:
            print(f"Could not apply Box-Cox to {col}: {str(e)}")
    else:
        # Shift data to make all values strictly positive
        shift = abs(df[col].min()) + 1  # Add 1 to ensure > 0
        try:
            transformed, _ = stats.boxcox(df[col] + shift)
            df[f'boxcox_{col}'] = transformed
            plot_transformation(df[col], df[f'boxcox_{col}'], f'Box-Cox shifted {col}')
        except ValueError as e:
            print(f"Could not apply Box-Cox to {col}: {str(e)}")

# Check skewness after transformation
for col in numeric_cols:
    print(f"\nSkewness for {col}:")
    print(f"Original: {df[col].skew():.2f}")
    
    if f'log_{col}' in df.columns:
        print(f"Log transform: {df[f'log_{col}'].skew():.2f}")
    
    if f'sqrt_{col}' in df.columns:
        print(f"Sqrt transform: {df[f'sqrt_{col}'].skew():.2f}")
    
    if f'boxcox_{col}' in df.columns:
        print(f"Box-Cox transform: {df[f'boxcox_{col}'].skew():.2f}")
```

```python
# For Box-Cox transformed columns (recommended as they often give best results)
for col in numeric_cols:
    if f'boxcox_{col}' in df.columns:
        df[col] = df[f'boxcox_{col}']  # Replace original with transformed
        df = df.drop(columns=[f'boxcox_{col}'])  # Remove the temp column

# For columns where Box-Cox failed, use log transform
for col in numeric_cols:
    if f'log_{col}' in df.columns and col in df.columns:  # If not already replaced
        df[col] = df[f'log_{col}']
        df = df.drop(columns=[f'log_{col}'])

# For columns where log failed, use sqrt transform
for col in numeric_cols:
    if f'sqrt_{col}' in df.columns and col in df.columns:  # If not already replaced
        df[col] = df[f'sqrt_{col}']
        df = df.drop(columns=[f'sqrt_{col}'])

# Verify the transformations
print(df[numeric_cols].skew())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Select numeric columns for outlier analysis (excluding binary/categorical columns)
numeric_cols = ['gf', 'xg', 'poss', 'sh', 'sot', 'dist', 'attendance']

# 1. Visual Outlier Detection with Boxplots
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numeric_cols])
plt.title('Boxplot of Numeric Columns (Before Outlier Treatment)')
plt.xticks(rotation=45)
plt.show()

# 2. Statistical Outlier Detection
def detect_outliers(df, columns, threshold=3):
    """
    Detect outliers using z-score method
    Returns a DataFrame with outlier flags
    """
    outlier_flags = pd.DataFrame(index=df.index)
    
    for col in columns:
        z_scores = np.abs(stats.zscore(df[col]))
        outlier_flags[f'outlier_{col}'] = z_scores > threshold
    
    return outlier_flags

# Detect outliers
outlier_flags = detect_outliers(df, numeric_cols)

# Show outlier counts per column
print("\nOutlier Counts per Column:")
print(outlier_flags.sum())

# Show sample of rows with outliers
print("\nSample Rows with Outliers:")
print(df[outlier_flags.any(axis=1)].head())

# 3. Visualize Outliers in Key Metrics
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.scatterplot(x='xg', y='gf', data=df, ax=axes[0, 0], hue=outlier_flags['outlier_xg'])
sns.scatterplot(x='sh', y='sot', data=df, ax=axes[0, 1], hue=outlier_flags['outlier_sh'])
sns.scatterplot(x='poss', y='dist', data=df, ax=axes[1, 0], hue=outlier_flags['outlier_poss'])
sns.scatterplot(x='attendance', y='gf', data=df, ax=axes[1, 1], hue=outlier_flags['outlier_attendance'])
plt.suptitle('Outlier Visualization in Key Metrics')
plt.tight_layout()
plt.show()
```

**Verification of Outlier Treatment**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# First, let's make sure we have our dataframe and numeric columns
# Assuming df is your DataFrame with the La Liga data
numeric_cols = ['gf', 'xg', 'poss', 'sh', 'sot', 'dist', 'attendance']

# 1. Outlier Detection Functions
def detect_outliers_zscore(df, columns, threshold=3):
    """Detect outliers using z-score method"""
    outlier_flags = pd.DataFrame(index=df.index)
    for col in columns:
        z_scores = np.abs(stats.zscore(df[col]))
        outlier_flags[f'outlier_{col}'] = z_scores > threshold
    return outlier_flags

def detect_outliers_iqr(df, columns, factor=1.5):
    """Detect outliers using IQR method"""
    outlier_flags = pd.DataFrame(index=df.index)
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - factor * IQR
        upper_bound = Q3 + factor * IQR
        outlier_flags[f'outlier_{col}'] = ~df[col].between(lower_bound, upper_bound)
    return outlier_flags

# 2. Outlier Treatment Functions
def winsorize_data(df, columns, lower=0.05, upper=0.95):
    """Cap outliers at specified percentiles"""
    df_clean = df.copy()
    for col in columns:
        lower_bound = df[col].quantile(lower)
        upper_bound = df[col].quantile(upper)
        df_clean[col] = df[col].clip(lower_bound, upper_bound)
    return df_clean

def iqr_outlier_treatment(df, columns, factor=1.5):
    """Cap values using IQR method"""
    df_clean = df.copy()
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - factor * IQR
        upper_bound = Q3 + factor * IQR
        df_clean[col] = df[col].clip(lower_bound, upper_bound)
    return df_clean

# 3. Visualize Outliers Before Treatment
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numeric_cols])
plt.title('Boxplot of Numeric Columns (Before Outlier Treatment)')
plt.xticks(rotation=45)
plt.show()

# 4. Detect Outliers (using both methods)
outliers_zscore = detect_outliers_zscore(df, numeric_cols)
outliers_iqr = detect_outliers_iqr(df, numeric_cols)

print("Outlier Counts (Z-score method):")
print(outliers_zscore.sum())

print("\nOutlier Counts (IQR method):")
print(outliers_iqr.sum())

# 5. Apply Outlier Treatment
# Create cleaned versions using different methods
df_winsorized = winsorize_data(df, numeric_cols)
df_iqr = iqr_outlier_treatment(df, numeric_cols)

# 6. Compare Results
def compare_outliers(original, treated, columns, method='zscore'):
    """Compare outlier counts before and after treatment"""
    if method == 'zscore':
        orig_outliers = detect_outliers_zscore(original, columns).sum()
        treated_outliers = detect_outliers_zscore(treated, columns).sum()
    else:  # iqr
        orig_outliers = detect_outliers_iqr(original, columns).sum()
        treated_outliers = detect_outliers_iqr(treated, columns).sum()
    
    comparison = pd.DataFrame({
        'Original': orig_outliers,
        'Treated': treated_outliers,
        'Reduction (%)': ((orig_outliers - treated_outliers) / orig_outliers) * 100
    })
    return comparison

print("\nOutlier Reduction - Winsorizing Method (Z-score detection):")
print(compare_outliers(df, df_winsorized, numeric_cols, 'zscore'))

print("\nOutlier Reduction - IQR Method (IQR detection):")
print(compare_outliers(df, df_iqr, numeric_cols, 'iqr'))

# 7. Visual Comparison
def plot_comparison(original, treated, cols_to_plot):
    plt.figure(figsize=(15, 6))
    plt.subplot(1, 2, 1)
    sns.boxplot(data=original[cols_to_plot])
    plt.title('Original Data')
    plt.xticks(rotation=45)
    
    plt.subplot(1, 2, 2)
    sns.boxplot(data=treated[cols_to_plot])
    plt.title('After Treatment')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Plot comparison for winsorized data
print("\nVisual Comparison - Winsorizing Method:")
plot_comparison(df, df_winsorized, ['xg', 'gf', 'attendance'])

# Plot comparison for IQR treated data
print("\nVisual Comparison - IQR Method:")
plot_comparison(df, df_iqr, ['xg', 'gf', 'attendance'])

# 8. Final Treatment Recommendation
# For football data, we'll use a combination approach:
df_final = df.copy()

# Winsorize performance metrics (preserve extreme performances but cap them)
perf_cols = ['xg', 'gf', 'sh', 'sot']
df_final[perf_cols] = winsorize_data(df, perf_cols)[perf_cols]

# IQR treatment for attendance and distance (more conservative for these)
other_cols = ['attendance', 'dist']
df_final[other_cols] = iqr_outlier_treatment(df, other_cols)[other_cols]

# Leave possession (poss) as is since 100% is theoretically possible

print("\nFinal Outlier Counts (Mixed Approach):")
print(compare_outliers(df, df_final, numeric_cols, 'zscore'))

# Visualize final result
print("\nFinal Treated Data Visualization:")
plot_comparison(df, df_final, numeric_cols)

# Now df_final contains your data with outliers properly treated
# You can continue your analysis with df_final
```

```python
df['overperformance'] = df['gf'] - df['xg']
```

```python
# Avoid division by zero
df['shot_accuracy'] = df['sot'] / df['sh'].replace(0, 1)
df['xg_per_shot'] = df['xg'] / df['sh'].replace(0, 1)
df['xg_per_sot'] = df['xg'] / df['sot'].replace(0, 1)
df['shots_per_poss'] = df['sh'] / df['poss'].replace(0, 1)
df['xg_diff'] = df['xg'] - df['xga']
features = ['poss', 'sh', 'sot', 'dist', 'attendance', 
            'shot_accuracy', 'xg_per_shot', 'xg_per_sot', 
            'shots_per_poss', 'xg_diff', 'result']
x = df[features]
y = df['overperformance']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in features:
    if df[col].dtype == 'object':
        x[col] = le.fit_transform(x[col])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")
```
